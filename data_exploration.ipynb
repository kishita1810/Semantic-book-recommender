### Step 3: Perform exploratory data analysis (EDA)

In the video they just use Pycharms inbuilt features to inspect the data. 
But I have written some code to get relevant descriptive statistics.

#### Here are the steps I followed:
1. Load the dataset

2. Check for missing values

3. Get summary statistics

4. Visualize key distributions


```python
#!pip install matplotlib seaborn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
books = pd.read_csv(f"{path}/books.csv")  # Update this with the actual file path

# Display basic information about the dataset
print("Dataset Info:")
print(books.info())

# Show the first few rows
print("\nFirst 5 rows of the dataset:")
print(books.head())

# Check for missing values
print("\nMissing Values:")
print(books.isnull().sum())

# Get summary statistics
print("\nDescriptive Statistics:")
print(books.describe(include="all"))

# Visualizing distributions
plt.figure(figsize=(12, 6))
sns.histplot(books["average_rating"], bins=20, kde=True)
plt.title("Distribution of Ratings")
plt.xlabel("Average Rating")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(books["num_pages"].dropna(), bins=30, kde=True)
plt.title("Distribution of Number of Pages")
plt.xlabel("Number of Pages")
plt.ylabel("Count")
plt.show()

# Count of books per genre (if genre column exists)
if "genre" in books.columns:
    plt.figure(figsize=(12, 6))
    books["genre"].value_counts().nlargest(10).plot(kind="bar")
    plt.title("Top 10 Most Common Genres")
    plt.xlabel("Genre")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```
