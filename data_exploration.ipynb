```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
books = pd.read_csv(f"{path}/books.csv")  # Update this with the actual file path

# Display basic information about the dataset
print("Dataset Info:")
print(books.info())

# Show the first few rows
print("\nFirst 5 rows of the dataset:")
print(books.head())

# Check for missing values
print("\nMissing Values:")
print(books.isnull().sum())

# Get summary statistics
print("\nDescriptive Statistics:")
print(books.describe(include="all"))

# Visualizing distributions
plt.figure(figsize=(12, 6))
sns.histplot(books["average_rating"], bins=20, kde=True)
plt.title("Distribution of Ratings")
plt.xlabel("Average Rating")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(books["num_pages"].dropna(), bins=30, kde=True)
plt.title("Distribution of Number of Pages")
plt.xlabel("Number of Pages")
plt.ylabel("Count")
plt.show()

# Count of books per genre (if genre column exists)
if "genre" in books.columns:
    plt.figure(figsize=(12, 6))
    books["genre"].value_counts().nlargest(10).plot(kind="bar")
    plt.title("Top 10 Most Common Genres")
    plt.xlabel("Genre")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```

subtitle is missing for 65% of the books, meaning it may not be a reliable feature.

Other missing values are minimal (<5%) and can be handled via imputation or removal.

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Colums")
plt.ylabel("Missing Values")

plt.show()
```

#### It is important to check whether missing values have a pattern because this can imply some sort of bias. Subtitles can be discarded as it has most missing values. Authors, categories, thumbnail, description, and published_year have random missing values. But average_rating, num_pages, and ratings_count have missing values in a similar pattern. This may be a source of bias in the dataset.

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman") # Spearman correlation is more suitable for non continuous values    

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap =sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman Correlation"})

heatmap.set_title("Correlation Heatmap")
plt.show()
```

#### What we are interested in here is the correlation of missing_description with other variables. For example, if correlation between missing_description and num_pages is high, it would indicate that either longer books or shorter books have more missing descriptions. And we are not going to use missing_description in our model, so the recommender can be biased towards longer or shorter books. Same for age_of_book and average_rating. But as we can see in the heatmap, these values are not very strong related to missing_description. (A value of 1 or -1 indicates a perfect positive or negative correlation, while a value closer to 0 indicates no correlation.)

```python
books[(books["description"].isna()) |
    (books["num_pages"].isna()) |
    (books["average_rating"].isna()) |
    (books["published_year"].isna())
]
```

We will be only losing 303 rows, which is under 5% considering the dataset we have so we can go ahead and remove these rows.

```python
book_missing = books[~(books["description"].isna()) &
    ~(books["num_pages"].isna()) &
    ~(books["average_rating"].isna()) &
    ~(books["published_year"].isna())
]

book_missing.info(10)
```

Categories is a mess. There are some categories which are way too specific and some which are way too general. We will group them into broader categories.

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
# Get category counts
category_counts = book_missing["categories"].value_counts().reset_index()
category_counts.columns = ["Category", "Count"]  # Rename columns
category_counts = category_counts.sort_values("Count", ascending=False)  # Sort in descending order

# Plot the bar chart
plt.figure(figsize=(15, 8))  # Increase figure size for readability
sns.barplot(data=category_counts, x="Category", y="Count", palette="viridis")

plt.xticks(rotation=90, ha="right")  # Rotate x-axis labels for better readability
plt.xlabel("Book Categories")
plt.ylabel("Number of Books")
plt.title("Book Counts by Category")
plt.tight_layout()  # Adjust layout to prevent label cutoff

plt.show()
```
